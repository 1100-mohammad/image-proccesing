```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("pic.jpg")
constent = cv2.copyMakeBorder(img,10,10,10,10,cv2.BORDER_CONSTANT,value = [0,0,255])
replicate = cv2.copyMakeBorder(img,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img,10,10,10,10,cv2.BORDER_REFLECT)
```

```python
plt.subplot(231),plt.imshow(img),plt.title("orginal")
plt.subplot(232),plt.imshow(constent),plt.title("constent")
plt.subplot(233),plt.imshow(replicate),plt.title("replicate")
plt.subplot(234),plt.imshow(reflect),plt.title("reflect")
plt.show()
```
